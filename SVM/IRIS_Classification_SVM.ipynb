```python
import pandas as p
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
#loading data frome
df  = p.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
#appending Target
df['target'] = iris.target
df
```

```python
iris.target_names
```

```python
#for displaying data having target as 1

df[df.target == 1]
```

```python
#for displaying data having target as 2

df[df.target == 2]
```

```python
#Assiging a new column to DataFrame
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'],axis='columns')
X
```

```python
Y = df.target
Y
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y,test_size=0.2)
```

```python
len(X_Train)
```

```python
len(X_Test)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_Train,Y_Train)
```

```python
#Model Accuracy
model.score(X_Test,Y_Test)
```
